```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-100, 100.5, 0.5)
y = np.cos(x)
z = np.sin(x)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='cosine(x)')
plt.plot(x, z, label='sine(x)')

plt.legend()

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine Functions')

plt.grid(True)
plt.show()
```
